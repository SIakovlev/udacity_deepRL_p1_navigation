```python
import seaborn as sns
import numpy as np
# sns.set(style="darkgrid")
import pandas as pd
import matplotlib.pyplot as plt
# %matplotlib inline
plt.style.use('seaborn-darkgrid')
from scipy.signal import savgol_filter
```

```python
ddqn_fname=r'C:\Study\p1_navigation\reports\simple\ddqn_agent__10_15_2018_02_51_PM.npy'
dqn_fname=r'C:\Study\p1_navigation\reports\simple\dqn_agent__10_15_2018_01_46_PM.npy'
dqn_per_fname=r'C:\Study\p1_navigation\reports\simple\dqn_agent_PER_10_15_2018_04_04_PM.npy'
ddqn_per_fname=r'C:\Study\p1_navigation\reports\simple\ddqn_agent_PER_10_15_2018_03_40_PM.npy'

data = {}

data['dqn'] = np.load(dqn_fname)
data['ddqn'] = np.load(ddqn_fname)
data['dqn_per'] = np.load(dqn_per_fname)
data['ddqn_per'] = np.load(ddqn_per_fname)

```

```python
palette = plt.get_cmap('Dark2')

plt.figure(figsize=(16,9))

for i, column in enumerate(['ddqn', 'dqn', 'dqn_per', 'ddqn_per']):
    smoothed = savgol_filter(data[column], 101, 1)
    plt.plot(smoothed[:-50], label=column, linewidth=2, color=palette(i))
    plt.plot(data[column], linewidth=1, color=palette(i), alpha=0.2)
    plt.legend()
    
plt.plot([13.0]*1000, linestyle='--', color=palette(10))
plt.show()
```
